```python
#loading our dataset

import pandas as pd

%sql DROP TABLE if EXISTS traffic_accidents;

%sql PERSIST traffic_accidents;

%sql SELECT * FROM traffic_accidents LIMIT 5;
```

```python
#1. Display a list of 5 makes that had the highest number of crashes?
%%sql  
SELECT MAKE,
  COUNT(MAKE) AS highest_make_crashes
FROM traffic_accidents
GROUP BY MAKE
ORDER BY COUNT(MAKE) DESC
LIMIT 5;
```

```python
#2. Most personal vehicles involved in crashes were traveling towards which direction?
%%sql
SELECT TRAVEL_DIRECTION,
  COUNT(TRAVEL_DIRECTION) AS passenger_travel_direction
FROM traffic_accidents  WHERE VEHICLE_TYPE = 'PASSENGER'
GROUP BY TRAVEL_DIRECTION
ORDER BY COUNT(TRAVEL_DIRECTION) DESC;

#Most personal vehicles were traveling towards the North
```

```python
#3. How many passengers were involved in personal cars were in accidents?
%%sql
SELECT SUM(NUM_PASSENGERS)
FROM traffic_accidents 
WHERE VEHICLE_USE = 'PERSONAL';
```

```python
#How many passengers were involved in accidents while traveling in personal Sport Utility vehicles?
%%sql

SELECT SUM(NUM_PASSENGERS)
FROM traffic_accidents
WHERE VEHICLE_TYPE = 'SPORT UTILITY VEHICLE (SUV)'
```

```python
#Compute the total number of Mercedez-Benz vehicles involved in accidents while being parked.
%%sql
SELECT COUNT(CRASH_RECORD_ID)
FROM traffic_accidents
WHERE MAKE = 'MERCEDES-BENZ' 
AND  MANEUVER = 'PARKED'
```

```python
#Compute the total occupant count per vehicle count.
%%sql
SELECT MAKE,
  SUM(OCCUPANT_CNT) as total_occupants
FROM traffic_accidents
GROUP BY MAKE
```

```python
#Sort the number of passengers in descending order by travel direction taking into
#account travel direction. Return travel direction and no. of passengers.

%%sql

SELECT *
from traffic_accidents 
ORDER BY OCCUPANT_CNT DESC
LIMIT 5
```

```python
#Select the top 5 vehicle models involved in the accident with no defect, order them by make, and show their maneuver.
%%sql
SELECT MODEL, MANEUVER,
  COUNT(MODEL) as top_models_in_accident
FROM traffic_accidents WHERE VEHICLE_DEFECT = 'NONE'
GROUP BY MODEL
ORDER BY COUNT(MODEL) DESC
LIMIT 5;
```

```python
#Which ridesharing vehicles in an accident had the most number of passengers?Return model, and the number of passengers.

%%sql
SELECT MODEL,
  SUM(OCCUPANT_CNT) AS total_passengers_ride_sharing
FROM traffic_accidents WHERE VEHICLE_USE = "RIDESHARE SERVICE"
GROUP BY MODEL
ORDER BY SUM(OCCUPANT_CNT) DESC
LIMIT 10;

```
